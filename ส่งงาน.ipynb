## 1.เขียนโปรแกรมเพื่อรัน แบบฝึกหัดที่ 1 และแบบฝึกหัดที่ 2

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def delete(self, key):
        if not self.head:
            return
        if self.head.data == key:
            self.head = self.head.next
            return
        current = self.head
        while current.next and current.next.data != key:
            current = current.next
        if current.next:
            current.next = current.next.next

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")
```

## 2.จงเขียนโปรแกรมเพื่อสร้าง Doubly Linked-List

```python
class Node:

    def __init__(self, data):
        self.data = data
        self.next = None  # ชี้ไปยังโหนดถัดไป
        self.prev = None  # ชี้ไปยังโหนดก่อนหน้า

class DoublyLinkedList:

    def __init__(self):
        self.head = None  # จุดเริ่มต้นของ Doubly Linked List

    def append(self, data):

        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node
        new_node.prev = current

    def prepend(self, data):

        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        new_node.next = self.head
        self.head.prev = new_node
        self.head = new_node

    def display_forward(self):

        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

    def display_backward(self):

        current = self.head
        if not current:
            print("None")
            return
        while current.next:
            current = current.next
        while current:
            print(current.data, end=" -> ")
            current = current.prev
        print("None")


dll = DoublyLinkedList()

dll.append(1)
dll.append(2)
dll.append(3)

dll.prepend(0)

print("แสดงข้อมูลจากหัวไปท้าย:")
dll.display_forward()

print("แสดงข้อมูลจากท้ายไปหัว:")
dll.display_backward()
```

## 3.จงเขียนโปรแกรมใช้ Linked List ในการจัดการข้อมูลนักศึกษา โดยมีความสามารถดังนี้:
## เพิ่มข้อมูลนักศึกษา (รหัส, ชื่อ, เกรด)
## ลบข้อมูลนักศึกษาตามรหัสนักศึกษา
## ค้นหาและแสดงข้อมูลนักศึกษาตามรหัสนักศึกษา
## แสดงรายชื่อนักศึกษาทั้งหมดเรียงตามรหัสนักศึกษา

```python
class StudentNode:

    def __init__(self, student_id, name, grade):
        self.student_id = student_id
        self.name = name
        self.grade = grade
        self.next = None  # ชี้ไปยังโหนดถัดไป

class StudentLinkedList:

    def __init__(self):
        self.head = None

    def add_student(self, student_id, name, grade):
        """เพิ่มข้อมูลนักศึกษาตามลำดับรหัส"""
        new_node = StudentNode(student_id, name, grade)
        if not self.head or self.head.student_id > student_id:
            # เพิ่มเป็นโหนดแรกใน List
            new_node.next = self.head
            self.head = new_node
            return
        # เพิ่มในตำแหน่งที่เหมาะสม
        current = self.head
        while current.next and current.next.student_id < student_id:
            current = current.next
        new_node.next = current.next
        current.next = new_node

    def delete_student(self, student_id):

        if not self.head:
            print(f"ไม่พบข้อมูลนักศึกษาที่มีรหัส {student_id}")
            return
        if self.head.student_id == student_id:
            # ลบโหนดแรก
            self.head = self.head.next
            print(f"ลบข้อมูลนักศึกษาที่มีรหัส {student_id} สำเร็จ")
            return
        current = self.head
        while current.next and current.next.student_id != student_id:
            current = current.next
        if not current.next:
            print(f"ไม่พบข้อมูลนักศึกษาที่มีรหัส {student_id}")
        else:
            current.next = current.next.next
            print(f"ลบข้อมูลนักศึกษาที่มีรหัส {student_id} สำเร็จ")

    def search_student(self, student_id):

        current = self.head
        while current:
            if current.student_id == student_id:
                print(f"รหัส: {current.student_id}, ชื่อ: {current.name}, เกรด: {current.grade}")
                return
            current = current.next
        print(f"ไม่พบข้อมูลนักศึกษาที่มีรหัส {student_id}")

    def display_all_students(self):

        if not self.head:
            print("ไม่มีข้อมูลนักศึกษา")
            return
        current = self.head
        print("รายชื่อนักศึกษา:")
        while current:
            print(f"รหัส: {current.student_id}, ชื่อ: {current.name}, เกรด: {current.grade}")
            current = current.next

students = StudentLinkedList()

students.add_student(1001, "Alice", "A")
students.add_student(1003, "Bob", "B")
students.add_student(1002, "Charlie", "A")

students.display_all_students()

students.delete_student(1003)

students.display_all_students()

students.search_student(1002)
students.search_student(1005)
```
